```python
df_cleandata = pd.read_csv('movie_data/movie_cleandata.csv')
df_cleandata.head()
```

```python
genre_list = set([])
t = float('NaN')
genre1 = df_cleandata['genre1'].unique().tolist()
genre2 = df_cleandata['genre2'].unique().tolist()
genre3 = df_cleandata['genre3'].unique().tolist()
genre_list.update(genre1,genre2,genre3)
genre_list = list(genre_list)
t = genre_list[0]
genre_list.remove(t)
print(genre_list)
```

```python
from collections import defaultdict

count_dict0 = defaultdict(lambda: 0)
count_dict1 = defaultdict(lambda: defaultdict(lambda: 0))
count_dict_f = defaultdict(lambda: defaultdict(lambda: 0))
count_dict_m = defaultdict(lambda: defaultdict(lambda: 0))
for index, row in df_cleandata.iterrows():
    for genre in genre_list:
        if (row.genre1 == genre or row.genre2 == genre or row.genre3 == genre):
            count_dict0[genre] += 1
            count_dict1[genre]['under18'] += int(row['avgr_u18'])
            count_dict1[genre]['18-29'] += int(row['avgr_18-29'])
            count_dict1[genre]['30-44'] += int(row['avgr_30-44'])
            count_dict1[genre]['above45'] += int(row['avgr_a45'])
            count_dict_f[genre]['under18'] += int(row['avgr_u18f'])
            count_dict_f[genre]['18-29'] += int(row['avgr_18-29f'])
            count_dict_f[genre]['30-44'] += int(row['avgr_30-44f'])
            count_dict_f[genre]['above45'] += int(row['avgr_a45f'])
            count_dict_m[genre]['under18'] += int(row['avgr_u18m'])
            count_dict_m[genre]['18-29'] += int(row['avgr_18-29m'])
            count_dict_m[genre]['30-44'] += int(row['avgr_30-44m'])
            count_dict_m[genre]['above45'] += int(row['avgr_a45m'])
print(count_dict1)
```

```python
results_list = []
for key1,values in count_dict1.items():
    key1,values = str(key1),dict(values)
#     print(key1, values)
    for key0,value in count_dict0.items():
        key0,value = str(key0),int(value)
#         print(key0, value)
        if key1==key0:
#             print(values.items())
            result = {k: v/value for k,v in values.items()}
#             print(result)
            results_list.append(result)

print(results_list)
pd_avgr_age = pd.DataFrame(results_list, index=genre_list)
pd_avgr_age
pd_avgr_age.to_json('movie_data/average_score_agegroup.json', orient='table')
```

```python
# count average rating  given by female
results_list_f = []
for key_f,values_f in count_dict_f.items():
    keyf_f,values_m = str(key_f),dict(values_f)
    for key0,value in count_dict0.items():
        key0,value = str(key0),int(value)
        if key_f==key0:
            result = {k: v/value for k,v in values_f.items()}
            results_list_f.append(result)

print(results_list_f)
pd_avgr_agef = pd.DataFrame(results_list_f, index=genre_list)
pd_avgr_agef
pd_avgr_agef.to_json('movie_data/average_score_agegroupf.json', orient='table')
```

### 探索director和actor关系

```python
df_relationship = df_cleandata[['director_name','actor_1_name','actor_2_name','actor_3_name']]
df_relationship.head()
```

```python
sum_num = df_relationship.shape[0]
num_dir = df_relationship['director_name'].unique().shape[0]
num_act1 = df_relationship['actor_1_name'].unique().shape[0]
num_act2 = df_relationship['actor_2_name'].unique().shape[0]
num_act3 = df_relationship['actor_3_name'].unique().shape[0]
print(sum_num, num_dir, num_act1, num_act2, num_act3)
```

导演节点

```python
nodes_director = df_relationship[['director_name', 'actor_1_name']].groupby('director_name').count().reset_index().reset_index()
nodes_director.rename(columns={'index':'category','director_name':'name','actor_1_name':'value'},inplace=True)
nodes_director['category'] = 0
nodes_director.head()
```

```python
nodes_director_json = nodes_director.to_json(orient='records')
print(nodes_director_json)
```

actor_1节点

```python
nodes_actor1 = df_relationship[['actor_1_name','actor_2_name']].groupby('actor_1_name').count().reset_index().reset_index()
nodes_actor1.rename(columns={'index':'category','actor_1_name':'name','actor_2_name':'value'},inplace=True)
nodes_actor1['category'] = 1
nodes_actor1.head()
```

actor_2节点

```python
nodes_actor2 = df_relationship[['actor_2_name','actor_3_name']].groupby('actor_2_name').count().reset_index().reset_index()
nodes_actor2.rename(columns={'index':'category','actor_2_name':'name','actor_3_name':'value'},inplace=True)
nodes_actor2['category'] = 2
nodes_actor2.head()
```

```python
nodes_test = pd.concat([nodes_director,nodes_actor1],axis=0)
nodes_test_json = nodes_test.to_json(orient='records')
print(nodes_test_json)
```

director-actor1 links

```python
df_dir_act1 = df_relationship[['director_name','actor_1_name']]
df_dir_act1.rename(columns={'director_name':'source','actor_1_name':'target'},inplace=True)
df_dir_act1.head()
```

```python
links_test_json = df_dir_act1.to_json(orient='records')
print(links_test_json)
```

```python
relationship_json = {}
relationship_json['nodes'] = nodes_test_json
relationship_json['links'] = links_test_json
print(relationship_json)
```
