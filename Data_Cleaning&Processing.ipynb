```python
df_cleandata = pd.read_csv('movie_data/movie_cleandata.csv')
df_cleandata.head()
```

```python
genre_list = set([])
t = float('NaN')
genre1 = df_cleandata['genre1'].unique().tolist()
genre2 = df_cleandata['genre2'].unique().tolist()
genre3 = df_cleandata['genre3'].unique().tolist()
genre_list.update(genre1,genre2,genre3)
genre_list = list(genre_list)
t = genre_list[0]
genre_list.remove(t)
print(genre_list)
```

```python
from collections import defaultdict

count_dict0 = defaultdict(lambda: 0)
count_dict1 = defaultdict(lambda: defaultdict(lambda: 0))
count_dict_f = defaultdict(lambda: defaultdict(lambda: 0))
count_dict_m = defaultdict(lambda: defaultdict(lambda: 0))
for index, row in df_cleandata.iterrows():
    for genre in genre_list:
        if (row.genre1 == genre or row.genre2 == genre or row.genre3 == genre):
            count_dict0[genre] += 1
            count_dict1[genre]['under18'] += int(row['avgr_u18'])
            count_dict1[genre]['18-29'] += int(row['avgr_18-29'])
            count_dict1[genre]['30-44'] += int(row['avgr_30-44'])
            count_dict1[genre]['above45'] += int(row['avgr_a45'])
            count_dict_f[genre]['under18'] += int(row['avgr_u18f'])
            count_dict_f[genre]['18-29'] += int(row['avgr_18-29f'])
            count_dict_f[genre]['30-44'] += int(row['avgr_30-44f'])
            count_dict_f[genre]['above45'] += int(row['avgr_a45f'])
            count_dict_m[genre]['under18'] += int(row['avgr_u18m'])
            count_dict_m[genre]['18-29'] += int(row['avgr_18-29m'])
            count_dict_m[genre]['30-44'] += int(row['avgr_30-44m'])
            count_dict_m[genre]['above45'] += int(row['avgr_a45m'])
print(count_dict1)
```

```python
results_list = []
for key1,values in count_dict1.items():
    key1,values = str(key1),dict(values)
#     print(key1, values)
    for key0,value in count_dict0.items():
        key0,value = str(key0),int(value)
#         print(key0, value)
        if key1==key0:
#             print(values.items())
            result = {k: v/value for k,v in values.items()}
#             print(result)
            results_list.append(result)

print(results_list)
pd_avgr_age = pd.DataFrame(results_list, index=genre_list)
pd_avgr_age
pd_avgr_age.to_csv('movie_data/average_score_agegroup.csv')
```

```python
# count average rating  given by female
results_list_f = []
for key_f,values_f in count_dict_f.items():
    keyf_f,values_m = str(key_f),dict(values_f)
    for key0,value in count_dict0.items():
        key0,value = str(key0),int(value)
        if key_f==key0:
            result = {k: v/value for k,v in values_f.items()}
            results_list_f.append(result)

print(results_list_f)
pd_avgr_agef = pd.DataFrame(results_list_f, index=genre_list)
pd_avgr_agef
```

```python
# budget vs. imdb_score
results_list = []
i = -1
for genre in genre_list:
    for index,row in df_cleandata.iterrows():
        count_dict_budget = {}
        if (row.genre1 == genre or row.genre2 == genre or row.genre3 == genre):
            i = i + 1
            count_dict_budget['index'] = i
            count_dict_budget['genre'] = genre
            count_dict_budget['movie_title'] = row['movie_title']
            count_dict_budget['imdb_score'] = row['imdb_score']
            count_dict_budget['budget'] = row['budget']
            results_list.append(count_dict_budget)
# print(results_list)
pd_collect_budget = pd.DataFrame(results_list)
pd_collect_budget[pd_collect_budget['genre']=='Crime']
```
